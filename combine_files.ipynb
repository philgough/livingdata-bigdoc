# Combine Files from the Living Dataset

This code takes the calculated SVM values from the living dataset and places them into csv files. 

```python
import pandas as pd
```

```python
import os
```

## Load the data

```python
df = pd.DataFrame(columns=['minute'], data=[i for i in range(10080)])
```

```python
df.shape
```

```python
input_dir = "SVM CSV"
```

## Combine into a single dataframe

```python
for in_f in os.listdir(input_dir):
    print(in_f)

    tdf = pd.read_csv(input_dir+'/'+in_f)
#     print(tdf.describe())
    df[in_f[:4]] = tdf['Mean SVM (g)']
```

```python
df.head()
```

## Export to CSV files

*output_full.csv* is all data

*output_narrow.csv* is all rows for 4 users

*output_sample.csv* is 100 rows for 4 users

```python
df.to_csv('output_full.csv', index=False)
```
